```python
img = '/content/drive/MyDrive/UFAZ_Capstone_project/data/yolov8_data/test/images/mask_735_png.rf.cdce52a752ecdb7f0f6f147b73decec2.jpg'
img = cv.imread(img)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
coor = result[0].boxes.xyxy
arr = img
coors = coor[0]
x1 = int((float(coors[0]) - float(coors[2]) / 2) )
y1 = int((float(coors[1]) - float(coors[3]) / 2) )
x2 = int((float(coors[0]) + float(coors[2]) / 2) )
y2 = int((float(coors[1]) + float(coors[3]) / 2) )

c1 = (int(coors[0]),  int(coors[1]))
c2 = (int(coors[2]), int(coors[3]))


new_arr = cv.rectangle(arr, c1, c2, color=(255, 0, 0), thickness=2)


plt.imshow(new_arr)
plt.show()
```
